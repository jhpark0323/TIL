## A. 데이터 전처리 - 데이터 읽어오기

```python
df = pd.read_csv('NFLX.csv')
df
```

```python
df.info()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df
```

```python
df.info()
```

```python
df_a = df[['Date', 'Open', 'High', 'Low', 'Close']]
df_a
```

## B. 데이터 전처리 - 2021년 이후의 종가 데이터 출력하기

```python
date2 = '2021-01-01'
date2_datetime = pd.to_datetime(date2)
df_a['Date'][0] >= date2_datetime
```

```python
date2 = '2021-01-01'
date2_datetime = pd.to_datetime(date2)

df_b = df_a[df_a['Date'] >= date2_datetime]
df_b
```

```python
# 그래프 그리기(가로축 : 날짜, 세로축 : 종가)
plt.plot(df_b['Date'], df_b['Close'])

# 그래프 제목
plt.title('NFIX Close Price')

# x축 레이블
plt.xlabel('Date')

# x 축 설정(회전시키기)
plt.xticks(rotation=45)

# y축 레이블
plt.ylabel('Close Price')

# 그래프 표시
plt.show()
```

## C. 데이터 분석 - 2021년 이후 최고, 최저 종가 출력하기

```python
df_b
```

```python
max_price = df_b['Close'].max()
print(max_price)
min_price = df_b['Close'].min()
print(min_price)
```

```python
print('최고 종가:', max_price)
print('최저 종가:', min_price)
```

## D. 데이터 분석 - 2021년 이후 월 별 평균 종가 출력하기

```python
df_b
```

```python
# 월별로 그룹화
df_mean = df_b.groupby(df_b['Date'].dt.to_period("M")).mean()
df_mean
```

```python
df_mean.info()
```

```python
# 월별로 그룹화 할 때 index를 period로 바꿨는데 그러면 plt의 축으로 못씀 -> timestamp로 변경
df_mean.index = df_mean.index.to_timestamp()
df_mean
```

```python
# 그래프 그리기(가로축 : 날짜, 세로축 : 종가)
plt.plot(df_mean.index, df_mean['Close'])

# 그래프 제목
plt.title('Monthly Average Close Price')

# x축 레이블
plt.xlabel('Date')

# x 축 설정(회전시키기)
plt.xticks(rotation=45)

# y축 레이블
plt.ylabel('Average Close Price')

# x 축을 날짜 형식으로 인식하도록 설정
plt.gca().xaxis_date()

# 그래프 표시
plt.show()
```
